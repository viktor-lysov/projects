## 1.1. Вывод и оценка общей информации по данным

Для удобства проведения идентичных операций над несколькими датафреймами в дальнейшем создадим словарь с названиями датафреймов.

```python
# Создаем словарь датафреймов
dfs = {'df_acquisition': df_acquisition,
            'df_company_and_rounds': df_company_and_rounds,
            'df_education': df_education,
            'df_people': df_people,
            'df_degrees': df_degrees,
            'df_fund': df_fund,
            'df_investment': df_investment}
```

Теперь пройдемся поочередно по созданным датафреймам. Взглянем на первичную информацию.

```python
print("Основная информация по датафреймам словаря 'dfs'")

# При помощи цикла переберем словарь 'dfs'
# и выведем основную информацию о содержащихся в нем датафреймах
for df_name, df in dfs.items():
    print(f"\n{'=' * 50}")
    print(f"Датафрейм '{df_name}'\n{'=' * 50}\n")
    print(df.info())
    print(df.head())
```

* Во всех датафреймах присутствует большое количество пропусков.
* Содержимое, в целом, соответствует названиям столбцов.   
* Датафрейм `df_company_and_rounds`, судя по представленным в нем данным, дублированию названия столбца-ключа **'company ID'** и **'company id'** и самому названию исходного файла датасета, состоит из двух таблиц: одна с данными о компаниях, вторая — с данными о раундах финансирования. Логично было бы разделить их на `df_company` и `df_rounds`.   
* В одном из датафреймов `df_degrees` в столбце **'object_id'** в данных стоит убрать префикс **'p:'** для возможности присоединения к другим датафреймам по этому идентификатору.   
* Столбцы с датами можно оптимизировать, приведя их к формату `datetime64`, некоторые столбцы, содержащие различные id в формате `float64` будет уместно перевести в целочисленный формат. В целом, все столбцы с типом `float64` можно оптимизиовать до `int64`.  * В датафрейме `df_fund` также много пропусков, даты можно привести к типу `datetime64`.

## 1.2 Проверка дубликатов

Проверим датафреймы на дубликаты. Сначала выполним нормализацию значений, чтобы убрать возможные неявные дубликаты, вызванные внесением одних и тех же значений разным регистром.

```python
# Создадим функцию, которая соберет в датафрейм количество уникальных значений в столбцах
# после нормализации снова посчитаем уникальные значения и найдем разницу,
# выявив количество неявных дубликатов, вызванных разницей регистров
def count_uniques(dfs_dict):
    # Сюда будем передавать названия датафреймов, столбцов и количество уникальных значений в столбцах
    report = []
    # Циклом переберем датафреймы в словаре 'dfs'
    for df_name, df in dfs_dict.items():
        # Вложенным циклом переберем названия столбцов каждого из датафреймов
        for col in df.columns:
            # Считаем количество уникальных значений в столбце
            unique_count = df[col].nunique(dropna=True)
            # Передеаем в список на вывод полученные значения
            report.append({
                'df_name': df_name,
                'column_name': col,
                'uniques_count': unique_count,
            })
    
    return pd.DataFrame(report)
```

```python
# Применяем функцию к словарю, сохраняем информацию в 'unique_stats'
unique_stats = count_uniques(dfs)
```

Информация по датафреймам в исходном состоянии собрана, можем переходить к нормализации.

```python
# Циклом перебираем датафреймы в словаре 'dfs'
for df_name, df in dfs.items():
    # Нас интересуют только текстовые столбцы,
    # поэтому сохраняем в список столбцы в типом данных 'object'
    text_cols = df.select_dtypes(include='object').columns
    for col in text_cols:
        dfs[df_name][col] = df[col].str.lower()
```

```python
# Применяем функцию к словарюю, сохраняем информацию после нормализации в 'unique_stats_normalized'
unique_stats_normalized = count_uniques(dfs)
```

```python
# Для избежания дублирования названий столбцов до и после нормализации, переименуем один из них
unique_stats_normalized = unique_stats_normalized.rename(columns={'uniques_count': 'uniques_count_normalized'})
```

```python
# Объединим датафреймы с количествами уникальных ззначений до и после нормализации
unique_stats_diff = unique_stats.merge(
    unique_stats_normalized,
    on=['df_name', 'column_name'],
    how='outer'
)

# Создадим столбец 'diff' с разницей значений до и после нормализации.
# Это будет количеством неявных дубликатов
unique_stats_diff['diff'] = unique_stats_diff['uniques_count'] - unique_stats_diff['uniques_count_normalized']

# Создадим столбец 'share' с долей неявных дубликатов в процентах
unique_stats_diff['share'] = round((unique_stats_diff['diff'] /
                                    unique_stats_diff['uniques_count']), 4) * 100
```

Все готово. Можем выводить получившиеся значения на экран.

```python
# Выведем только те строки, где встретились дубликаты, в них значение 'diff' будет больше 0
unique_stats_diff[unique_stats_diff['diff'] > 0].sort_values(by=['diff', 'df_name', 'column_name'], ascending=False)
```

```python
print(f"Количество столбцов, в которых были дубликаты: {len(unique_stats_diff[unique_stats_diff['diff'] > 0])}")
print(f"Всего удалено неявных дубликатов: {unique_stats_diff[unique_stats_diff['diff'] > 0]['diff'].sum()}")
print(f"Дубликаты встретились в датафреймах: {unique_stats_diff[unique_stats_diff['diff'] > 0]['df_name'].unique()}")
```

Неявные дубликаты при нормализации удалились, их оказалось не так много: в основном менее 1% от исходных данных. Но в 5 с толбцах их было от 2 до почти 6%. Больше всего таких дублей нашлось в столбцах с именами и фамилиями датафрейма `df_people`.

Перейдем к оптимизации типов данных. Сначала приведем столбцы с датами к типу `datetime64`.

```python
# В словарь запишем датафреймы и названия столбцов, у которых будем менять тип данных
convert_to_datetime = {'df_acquisition': 'acquired_at',
                       'df_company': 'founded_at',
                       'df_rounds': 'funded_at',
                       'df_education': 'graduated_at',
                       'df_fund': 'founded_at',
                       'df_company_and_rounds': ['founded_at', 'funded_at']
                      }
# Напишем функцию, в которую передадим словарь со всеми датафреймами и словарь-маску, по которой
# в исходном словаре будут выбираться необходимые датафреймы, а затем столбцы в них
def convert_to_datetime_func(dict, map):
    # Цикл переберет словарь-маску
    for df_name, column in map.items():
        # Проверяем, присутствует ли ключ словаря-маски в исходном словаре 
        if df_name in dict:
            # Датафрейм 'df' будем редактировать далее, передаем в него значение из исходного словаря
            df = dict[df_name]
            # Чтобы в одном датафрейме можно было изменить несколько столбцов, их нужно передать в список
            columns_list = [column] if isinstance(column, str) else column
            # Перебираем значения, переданные в список
            for col in columns_list:
                # Проверяем, присутствует ли значение из списка среди столбцов редактируемого датафрейма
                if col in df.columns:
                    # Приводим столбец к типу datetime, в случае ошибки будет пропуск
                    df[col] = pd.to_datetime(df[col], errors='coerce')
                    print(f"Столбец '{col}' датафрейма '{df_name}' конвертирован в datetime64")
                else:
                    print(f"Столбец '{col}' в датафрейме '{df_name}' не найден")
```

```python
# Применим функцию к словарю датафреймов
convert_to_datetime_func(dfs, convert_to_datetime)
```

## 2.1. Раунды финансирования по годам

Составим сводную таблицу по годам в которой выделим:
* типичный размер средств, выделяемый в рамках одного раунда;
* общее количество раундов финансирования за этот год.

```python
# Копируем данные для построения сводной таблицы
df_pivot = df_company_and_rounds
```

```python
# Добавим столбец 'year' для группирования по годам
df_pivot['year'] = df_pivot['founded_at'].dt.year

# Добавим столбец 'month' для группирования по месяцам
df_pivot['month'] = df_pivot['founded_at'].dt.month
```

```python
df_pivot
```

```python
# Сгруппируем данные по году и создадим 2 вычисляемых столбца:
# с медианной суммой финансирования 'typical_raised_amount'
# и с количеством раундов финансирования 'rounds_count'
pivot_raised_amount = df_pivot.groupby('year').agg(
    typical_raised_amount=('raised_amount', 'median'),
    rounds_count=('funding_rounds', 'sum')
)

# Фильтруем данные, оставляя только годы с более чем 50 раундами финансирования
pivot_raised_amount_filtered = pivot_raised_amount[pivot_raised_amount['rounds_count'] > 50].reset_index()

# Создаем столбец с типичным размером средств, которые стартапы получали в рамках одного раунда финансирования
pivot_raised_amount_filtered['typical_raised_amount_share'] = round(pivot_raised_amount_filtered['typical_raised_amount'] / pivot_raised_amount_filtered['rounds_count'], 2)
```

```python
# Приведем столбцы к целочисленному типу, чтобы отбросить ненужные нули
for col in pivot_raised_amount_filtered.columns:
    pivot_raised_amount_filtered[col] = pivot_raised_amount_filtered[col].astype('int64')
```

```python
pivot_raised_amount_filtered.sort_values(by='year')
```

```python
plt.figure(figsize=(14, 8))

# Постриоим линейный график, отображающей динамику размера финансирования по годам
plt.plot(pivot_raised_amount_filtered['year'],
         pivot_raised_amount_filtered['typical_raised_amount_share'],
         marker='o'
        )
plt.title('Динамика типичного размера финансирования стартапов по годам')
plt.xlabel('Год')
plt.ylabel('Типичная сумма финансирования')
plt.grid(True, alpha=0.5)
plt.show()
```

Выведем информацию о том, в каком году типичный размер собранных в рамках одного раунда средств был максимален.

## 3.3. Исследование компаний с суммами сделок в 0-1 доллар


* Исследуем компании, которые были проданы за 0 или за 1 доллар, и при этом известно, что у них был ненулевой общий объём финансирования.

* Рассчитаем верхнюю и нижнюю границу выбросов для столбца `funding_total` и определим, каким процентилям границы соответствуют.

Продолжим работу с ранее сгруппированным по `company_id` датафреймом `df_cnr_grouped`. Также потребуется датафрейм `df_acquisition` с информацией о покупках компаний.   
Объеденим их по столбцам `company_id` и `acquired_company_id`, отфильтруем полученные значения, оставив только те компании, сумма сделки у которых составила  равна `1$` или равна `0`.

```python
# Объединим данные по идентификатору компании для сгруппированного датафрейма 'df_cnr_grouped'
# и по идендификатору покупаемой компании 'acquired_company_id' для датафрейма 'df_acquisition'
df_free = pd.merge(df_cnr_grouped,
                   df_acquisition,
                   how='left',
                   left_on='company_id',
                   right_on='acquired_company_id')

# Оставим только те компании, которые были куплены за 1 доллар ('price amount ')или бесплатно
# и при том, что у них ненулевой общий объем финансирования 'funding_total'
df_free = df_free[(df_free['price_amount'] <= 1) &
                  (df_free['funding_total'] > 0)]
```

```python
Q1_free = df_free['funding_total'].quantile(0.25)

Q3_free = df_free['funding_total'].quantile(0.75)

IQR_free = Q3_free - Q1_free

lower_free = Q1_free - 1.5 * IQR_free

upper_free = Q3_free + 1.5 * IQR_free

lower_free_percentile = round((df_cnr_grouped['funding_total'] < lower_free).mean() * 100, 2)
upper_free_percentile = round((df_cnr_grouped['funding_total'] > upper_free).mean() * 100, 2)
```

Нижняя граница выбросов получилась отрицательно, но финансирование не может быть отрицательным, а при рассматриваемых условиях оно не может быть и нулевым. Соответственно, в нижней части выбросов нет.
